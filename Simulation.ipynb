```python
import math
from statistics import mean,variance

def mean_numbers(dim, q, steps):
    n_roots = []
    n_iter = []
    for i in range(steps):
        W = wilson_RSF_grid_2d((dim,dim), q, False)
        n_roots.append(len(W[3]))
        n_iter.append(len(W[1]))
    print("==============================================")
    print("PARAMETERS: %dx%d grid, q = %f, steps = %d" %(dim,dim,q,steps))
    print("-------------------------------")
    print("MEAN NUMBER OF NODES: %d" %(mean(n_roots)))
    print("MSD: %d" %(math.sqrt(variance(n_roots))))
    print("-------------------------------")
    print("MEAN NUMBER OF ITERATIONS: %d" %(mean(n_iter)))
    print("MSD: %d" %(math.sqrt(variance(n_iter))))
    
    return
```

```python
mean_numbers(8, 0.3, 1000)
mean_numbers(8, 0.2, 1000)
mean_numbers(8, 0.1, 1000)
mean_numbers(8, 0.05, 1000)
```

```python
mean_numbers(10, 0.3, 1000)
mean_numbers(10, 0.2, 1000)
mean_numbers(10, 0.1, 1000)
mean_numbers(10, 0.05, 1000)
```
